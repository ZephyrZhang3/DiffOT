```python
from sklearn.datasets import make_swiss_roll

# 生成瑞士卷数据
X, colors = make_swiss_roll(n_samples=1500, noise=0.5, random_state=42)
print(f"{colors = }")
# 将数据转换为matplotlib的figure和axes对象
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# 绘制数据点
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=colors, cmap="rainbow")

# 设置轴标签
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# 显示图表
plt.show()
```
